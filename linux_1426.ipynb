문제4. 입력한 패턴과 일치하는 모든 경로에 대해 데이터베이스를 검색하여 표준 출력으로 다시 제공하는 명령어를 작성해주세요

정답 : locate <패턴> | grep <검색할_패턴>

* | : Shift+\(엔터키 위에 있는 키)로 입력

image.png

 #* (애스터리스크): 임의의 문자열을 의미함 0개 이상의 문자에 일치
.txt : 확장자가 .txt인 파일을 의미
 #*.txt : .txt 확장자를 가진 모든 파일을 의미
| (파이프) : 리눅스에서 왼쪽 명령어의 출력을 오른쪽 명령어의 입력으로 넘기는 역할

locate "*.txt" | grep "my" 에서 출력된 경로들이 |를 통해 grep "my" 명령어로 전달되고, grep은 그 중 "my"가 포함된 경로만 출력

###grep은 텍스트 패턴을 검색해서 일치하는 부분을 찾는 유틸
